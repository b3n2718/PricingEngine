```python
from scipy.stats import qmc
```

```python
import numpy as np
```

```python
n_steps = 252
n_sims = 100
total_noise_dim = 2
```

```python
d = n_steps * total_noise_dim
```

```python
sampler = qmc.Sobol(d=d, scramble=True)
uniform = sampler.random(n_sims)
```

```python
uniform.shape
```

```python
samples = np.clip(uniform, 1e-10, 1 - 1e-10)
```

```python
samples.reshape(n_sims, n_steps, total_noise_dim).shape
```
